**Binning**

```python
import matplotlib.pyplot as plt
plt.hist(dataset["Price"])
plt.xlabel("Price")
plt.ylabel("count")
plt.title("Price")
```

```python
# Rearrange the values into three bins':
bins = np.linspace(min(dataset["Price"]), max(dataset["Price"]), 4)
```

```python
bins
```

```python
# Set group names:

group_names = ['Low', 'Medium', 'High']
```

```python
# Determine what each value of dataset['Price'] belongs to:

dataset['Price-binned'] = pd.cut(dataset['Price'], bins, labels=group_names, include_lowest=True )
```
